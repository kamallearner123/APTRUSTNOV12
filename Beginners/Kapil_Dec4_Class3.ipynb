```python
# How to read a file .txt
file_name = "news.txt"

# 1) Read the data
# Open file and get File Descritor
fd = open(file_name)
data = fd.read() #Reading 10 bytes 
fd.close()

# 2) Split the data into words
# Get list and set
data = data.replace("\n", " ")
list_words = data.split(" ") # Delimiter  " "
words = set(list_words)

# 3) Create a dictionary
freq = {} # Empty

# 4) Take a word from set: Find count in list: Add "word":count
for word in words:
    # Check the word in list.
    #print(f"Word '{word}' repeated for {list_words.count(word)}")
    freq[word] = list_words.count(word)

# 5) Filter the words
# 6) Write result into a file result.txt
fd = open("result.txt", "w") # w - write
result = {}
for i in freq:
    if freq[i] >= 2:
        result[i] = freq[i]
        fd.write(f"{i}, {freq[i]}\n")
fd.close()

fd = open("result.txt") #read
data = fd.read()
print(data)
fd.close()
```

```python
x = 2
def solve_equation(*nums): # nums collects all args as list
    result = 0
    pow = 0
    for num in nums[::-1]:
        result += num*(x**pow)
        pow += 1
    return result
    
print(solve_equation(10,20))# 10x+20
print(solve_equation(1,10,20))# x**2+10x+20
# solve_equation(1,0,10)# x**2+10
```

```python
def solve_equation(x, *nums): # nums collects all args as list
    result = 0
    pow = 0
    for num in nums[::-1]:
        result += num*(x**pow)
        pow += 1
    return result
    
print(solve_equation(2, 10,20))# 10x+20
print(solve_equation(3, 1,10,20))# x**2+10x+20
# solve_equation(1,0,10)# x**2+10
```
